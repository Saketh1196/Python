```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'From_To':['LoNDon_paris','MAdrid_miLAN','londON_Stockholm','Budapest_PaRis','Brussels_LondOn'],
                   'FlightNumber':[10045,np.nan, 10065, np.nan, 10085],
                   'RecentDelays':[[23,47],[],[24,43,87],[13],[67,32]],
                   'Airline':['KLM(!)','<Air France>(12)','(British Airways. )','12. Air France',"'Swiss Air'"]})
df
```

```python
df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)
df
```

```python
Temp_DataFrame[['From','To']]= pd.DataFrame(df['From_To'].str.split("_",expand=True))
df = pd.concat([df,Temp_DataFrame[['From','To']]],axis=1)
df
```

```python
df['From']=df['From'].str.capitalize()
df['To']=df['To'].str.capitalize()
df
```

```python
Delays = pd.DataFrame(df['RecentDelays'].to_list(),columns=['delay_1','delay_2','delay_3'])
df = pd.concat([df,Delays],axis=1)
df.drop(columns='RecentDelays')
```
